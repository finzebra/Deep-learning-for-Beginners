```python
import numpy as np
import matplotlib.pyplot as plt

def plot_rnn_unrolling():
    time_steps = 10
    input_size = 1  # Scalar input at each time step
    hidden_size = 5  # Number of units in the hidden state
    
    # Generate some example input data (random values)
    input_data = np.random.randn(time_steps, input_size)
    
    # Initialize the hidden state and output
    hidden_state = np.zeros((time_steps, hidden_size))
    
    # Define simple weights for input and hidden state (random for visualization)
    w_input = np.random.randn(input_size, hidden_size)
    w_hidden = np.random.randn(hidden_size, hidden_size)
    
    # Calculate the hidden states over time (forward pass of an RNN)
    for t in range(time_steps):
        if t == 0:
            hidden_state[t] = np.tanh(np.dot(input_data[t], w_input))  # First step
        else:
            hidden_state[t] = np.tanh(np.dot(input_data[t], w_input) + np.dot(hidden_state[t-1], w_hidden))  # Subsequent steps
    
    # Plotting
    fig, ax = plt.subplots(figsize=(10, 6))
    for t in range(time_steps):
        ax.plot(hidden_state[t], label=f'Time Step {t+1}')
    
    ax.set_title('Evolution of Hidden States Over Time')
    ax.set_xlabel('Hidden Units')
    ax.set_ylabel('Hidden State Value')
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()

# Call the function to visualize
plot_rnn_unrolling()
```

```python
import plotly.graph_objects as go

def plot_rnn_flow():
    time_steps = 10
    input_size = 1
    hidden_size = 5
    
    input_data = np.random.randn(time_steps, input_size)
    hidden_state = np.zeros((time_steps, hidden_size))
    
    w_input = np.random.randn(input_size, hidden_size)
    w_hidden = np.random.randn(hidden_size, hidden_size)
    
    for t in range(time_steps):
        if t == 0:
            hidden_state[t] = np.tanh(np.dot(input_data[t], w_input))
        else:
            hidden_state[t] = np.tanh(np.dot(input_data[t], w_input) + np.dot(hidden_state[t-1], w_hidden))
    
    fig = go.Figure()

    for t in range(time_steps):
        fig.add_trace(go.Scatter(x=np.full(hidden_size, t+1), y=hidden_state[t], mode='markers', name=f'Time Step {t+1}', marker=dict(size=12)))
    
    fig.update_layout(
        title='Flow of Information in RNN Over Time',
        xaxis_title='Time Step',
        yaxis_title='Hidden State Value',
        showlegend=True
    )
    fig.show()

# Call the function to visualize
plot_rnn_flow()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Function to animate the evolution of hidden states
def animate_rnn():
    time_steps = 10
    input_size = 1
    hidden_size = 5
    
    input_data = np.random.randn(time_steps, input_size)
    hidden_state = np.zeros((time_steps, hidden_size))
    
    w_input = np.random.randn(input_size, hidden_size)
    w_hidden = np.random.randn(hidden_size, hidden_size)
    
    for t in range(time_steps):
        if t == 0:
            hidden_state[t] = np.tanh(np.dot(input_data[t], w_input))
        else:
            hidden_state[t] = np.tanh(np.dot(input_data[t], w_input) + np.dot(hidden_state[t-1], w_hidden))

    fig, ax = plt.subplots(figsize=(10, 6))
    lines = [ax.plot([], [], label=f"Unit {i+1}")[0] for i in range(hidden_size)]
    ax.set_xlim(0, time_steps - 1)
    ax.set_ylim(-1, 1)
    ax.set_title('Animated Hidden State Evolution')
    ax.set_xlabel('Time Step')
    ax.set_ylabel('Hidden State Value')
    ax.legend(loc='upper left')

    def init():
        for line in lines:
            line.set_data([], [])
        return lines

    def update(frame):
        for i, line in enumerate(lines):
            line.set_data(range(frame+1), hidden_state[:frame+1, i])
        return lines

    ani = FuncAnimation(fig, update, frames=range(time_steps), init_func=init, blit=True, interval=500)

    # Display the animation
    return HTML(ani.to_jshtml())

# Call the function to create the animation and display it
animate_rnn()
```
